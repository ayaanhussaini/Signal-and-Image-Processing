### Thresholding


```python
# importing image
img2= cv2.imread('Pollen.tif', 0)
```

```python
# finding bit size
print(img2.dtype)
```

```python
# showing image in greyscale
plt.imshow(img2, cmap = 'gray')
```

```python
# finding number of rows and columns
img2.shape
```

```python
# making a copy of image
img2new = img2.copy()
```

```python
threshold = 100  # threshold value

# looping through the image
for i in range (889):
  for j in range(889):
    if img2[i,j] < threshold:  # checking if pixel level is less than threshold
      img2new[i,j] = 0   # changing pixel level to 0
    else:
      img2new[i,j] = 255   # changing pixel level to 255
```

```python
# showing image in greyscale
plt.imshow(img2new, cmap = 'gray')
```

### Grey level slicing without background



```python
# importing image
img3= cv2.imread('Fig0312(a)(kidney).tif', 0)
```

```python
# finding number of rows and columns
img3.shape
```

```python
# finding bit size
print(img3.dtype)
```

```python
# showing image in greyscale
plt.imshow(img3, cmap = 'gray')
```

```python
# making a copy of image
img3_without_bg = img3.copy()
```

```python
#  defining range
lower = 140
upper = 200

# looping through the image
for i in range(828):
  for j in range(720):
    if img3_without_bg[i,j] < upper and img3_without_bg[i,j] > lower:    # checking if pixel level is in the range
      img3_without_bg[i,j] = 255   # changing value in the range to 255
    else:
       img3_without_bg[i,j]= 0  # changing value of background to 0
```

```python
# showing image in greyscale
plt.imshow(img3_without_bg, cmap = 'gray')
```

### Low Pass Filtering

```python
# making a copy of image
img4= img1.copy()
```

```python
# finding number of rows and columns
img4.shape
```

```python
mask_size = 7   # mask size
a = mask_size//2   # number of pixels on each side of teh origin

# looping through the image
for i in range (a,511):
  for j in range(a, 511):
    temp = img4[i-a:i+a+1, j-a:j+a+1]

    sum = np.sum(temp)  # sum of values in mask
    img4[i,j]= sum//mask_size**2    # sum/ number of pixels in mask

```

```python
# showing image in greyscale
plt.imshow(img4, cmap = 'gray')
```
