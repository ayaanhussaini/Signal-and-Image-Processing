```python
# plotting sine wave 

# x = Asin(2*pi*f*t)

t = np.arange(0,1,0.01)
f= 2
A = 1
x = A*np.sin(2 * np.pi * f * t)
y = A*np.cos(2 * np.pi * f * t)
```

```python
plt.plot(t,x, "r*-", t,y)
plt.xlabel('time')
plt.ylabel('amplitude')
plt.title('Sine and Cosine Waves')
```

```python
plt.subplot(1,3,1)
plt.plot(t,x)
plt.xlabel('time')
plt.ylabel('amplitude')
plt.title('Sine Wave')
plt.subplot(1,3,3)
plt.plot(t,y)
plt.xlabel('time')
plt.ylabel('amplitude')
plt.title('Cosine Wave')
```

```python
# making a discrete wave
plt.stem(t,x)
plt.xlabel('time')
plt.ylabel('amplitude')
plt.title('Sine Wave')
```
