# Spatial Resolution- Down-sampling 

```python
# Importing libraries
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#reading the original image and know its type
img1= cv2.imread('cameraman.tif',0)
type(img1)
```

```python
# finding max pixel level
np.max(img1)
```

```python
# finding the size of the original image
[m,n]=img1.shape
print(m,n)
```

```python
# getting sampling rate from the user
factor= int(input("Enter the down sampling rate: " ))
```

```python
# reducing the new image using slicing(by skipping rows and columns)
img2 = img1[::factor,::factor]
```

```python
#finding resolution of the new down sampled image
img2.shape
```

```python
plt.figure(figsize=(12,12))

#display original image
plt.subplot(1,2,1)
plt.imshow(img1, cmap='gray')
#displau down sampled image
plt.subplot(1,2,2)
plt.imshow(img2,cmap='gray')
```

# Grey level resolution

```python
# finding maximum level 
max_old= np.max(img1)
max_old
```

```python
# finding number of rows and colunms and assigning them to variables
[m,n]=img1.shape
print(m,n)
```

```python
#  user inputting the number of bits for the new image
b= int(input("Enter the number of bits for the new image: "))
```

```python
#Calculate the constant value to do the mapping
max_new= (2**b)-1
constant = (max_new)/(max_old)
constant
```

```python
# making a new image
img3= np.zeros((m,n))

# applying the constant to the image
img3 = (img1*constant)
img3= np.round(img3)
```

```python
plt.figure(figsize=(12,12))

#Show original image
plt.subplot(1,2,1)
plt.imshow(img1, cmap='gray')

#Show new image
plt.subplot(1,2,2)
plt.imshow(img3,cmap='gray')
```
