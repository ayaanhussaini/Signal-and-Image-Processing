```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/content/Fig0911(a)(noisy_fingerprint).tif",0)
plt.imshow(img, cmap = 'gray', vmin= 0,vmax=256)
print(img.dtype)
print(img.shape)
```

```python
m,n= img.shape
```

### Opening

```python
def erosion(eroding, k):
  k= k

  SE= np.ones((k,k)) 
  const= (k-1)//2

  eroded_img= np.zeros((m,n))


  for i in range(const, m-const):
    for j in range(const, n-const):
      temp= eroding[(i-const):(i+const+1), (j-const):(j+const+1)]
      product= temp*SE
      eroded_img[i,j]= np.min(product)
  return(eroded_img)


eroded = erosion(img, 3)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(img, cmap="gray")
plt.title("Original")
plt.subplot(1,2,2)
plt.imshow(eroded, cmap="gray")
plt.title("Eroded")
```

```python
def dialate(dialating, k):
  k= k

  SE= np.ones((k,k)) 
  const= (k-1)//2

  dialated_img= np.zeros((m,n))


  for i in range(const, m-const):
    for j in range(const, n-const):
      temp= dialating[(i-const):(i+const+1), (j-const):(j+const+1)]
      product= temp*SE
      dialated_img[i,j]= np.max(product)
  return(dialated_img)
```

```python
eroded.shape
```

```python
opening = dialate(eroded, 3)

plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(eroded, cmap="gray")
plt.title("Eroded")
plt.subplot(1,2,2)
plt.imshow(opening, cmap="gray")
plt.title("Opening")
```

### Closing

```python
def dialate(dialating, k):
  k= k

  SE= np.ones((k,k)) 
  const= (k-1)//2

  dialated_img= np.zeros((m,n))


  for i in range(const, m-const):
    for j in range(const, n-const):
      temp= dialating[(i-const):(i+const+1), (j-const):(j+const+1)]
      product= temp*SE
      dialated_img[i,j]= np.max(product)
  return(dialated_img)

dialated = dialate(opening, 3)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(opening, cmap="gray")
plt.title("Opening")
plt.subplot(1,2,2)
plt.imshow(dialated, cmap="gray")
plt.title("Dialated after opening")
```

```python
def erosion(eroding, k):
  k= k

  SE= np.ones((k,k)) 
  const= (k-1)//2

  eroded_img= np.zeros((m,n))


  for i in range(const, m-const):
    for j in range(const, n-const):
      temp= eroding[(i-const):(i+const+1), (j-const):(j+const+1)]
      product= temp*SE
      eroded_img[i,j]= np.min(product)
  return(eroded_img)


closing = erosion(dialated, 3)


plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(dialated, cmap="gray")
plt.title("Dialated after opening")
plt.subplot(1,2,2)
plt.imshow(closing, cmap="gray")
plt.title("Closing")
```
