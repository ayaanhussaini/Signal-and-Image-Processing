```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

## Ramp signal

```python
n = np.arange(-10,11)# define discrete time signal
r = np.where(n>= 0, n, 0) # ramp r=n  for r>0
print(r)
plt.stem(n,r)
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

## Unit step

```python
n = np.arange(-10,11)# define discrete time signal
r = np.where(n>=0, 1, 0)
print(r)
plt.stem(n,r)
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

## Impuse signal

### Time shifting

```python
# plotting a wave
n = np.arange(-4,6,1)
x = np.array([0,0,0,0,2,1,2,1,0,0])
plt.title('Original')
plt.stem(n,x)
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

```python
# time reversal
y = np.flipud(x)
plt.stem(n,y)
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

## Amplification

```python
# y(n) = 2x(n)
amp = x*2
print(amp)
plt.stem(n,amp)
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

## Attenuation

```python
# y(n) = x(n)/2
att = x/2
print(att)
plt.stem(n,att)
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

## Time scaling

```python
# plotting a wave
x= np.array([1,2,3,4,5,4,3,2,1])
n= np.arange(0,9,1)


print(n)
print(x)
plt.title('Original') 
plt.stem(n,x)
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

```python
# (y(n)= x(2n)---> compress

x= np.array([1,2,3,4,5,4,3, 2,1])
n= np.arange(0,9,1)

y=[]

for i in x:
  if i%2==0:
    y.append(i)


print(n)
print(y)

plt.stem(y)
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

```python
# (y(n)= x(n/2)---> interpolation 

x= np.array([1,2,3,4,5,4,3,2,1])
n= np.arange(0,9,1)

y=[]

for i in x:
    y.append(i)
    y.append(0)


n2= []
len = (n.size)*2
for j in range(len):
    n2.append(j)


print(n2)
print(y)

plt.stem(n2,y)
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

## Converting Continuous Signals to Discrete

```python
# making a sine wave
t = np.arange(0,1,0.05)
f= 1
A = 1
x = A*np.sin(2 * np.pi * f * t)
```

```python
x
```

```python
# Continuous Time and Continuous Amplitude Sine Wave
plt.plot(t,x)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Continuous Time and Continuous Amplitude Sine Wave')
```

```python
# making time discrete
plt.stem(t,x)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Continuous Time and Discrete Time Sine Wave')
```

```python
# Making Amplitude discrete 
x1 = np.round(x,decimals = 0)
x1
```

```python
# Plotting Discrete Amplitude and Discrete Time wave
plt.stem(t,x1)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Discrete Amplitude and Discrete Time Sine Wave')
```

```python
# Arranging them together to see the difference 

plt.figure(figsize = [22,5])

plt.subplot(1,3,1)

plt.plot(t,x)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Continuous Time and Continuous Amplitude Sine Wave')

plt.subplot(1,3,2)
plt.stem(t,x)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Continuous Time and Discrete Time Sine Wave')

plt.subplot(1,3,3)
plt.stem(t,x1)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Discrete Amplitude and Discrete Time Sine Wave')
```
